# Conditional Statement

### if else statement

```python
no = 12

if no > 0 : 
    print ("No is Positive")
else :
    print ("No is Negative")
```

### if elif else Statement

```python
no = 0

if no == 0 :
    print ("NO Is Zero")
elif no > 0 :
    print ("NO is Positive")
else :
    print ("NO is Negative")
```

```python
age = 16

if age > 100 :
    print ("Not a Human")
elif age > 22 : 
    print ("we can Marage")
elif age > 18 : 
    print ("You are Adult ")
else : 
    print ("AA MERE BACHE FANTA PIYE Ga") 
```
