```python
class A :
    c = "Class : A"

    def __len__ (self) :
        return 10

    def __str__ (self) :
        return "String"

    @classmethod
    def Fun1 (cls):
        return cls.c

    @staticmethod
    def Fun2 () :
        return "Static Function"
        

ob = A ()

print ( str (ob) )
print ( len (ob) )
print ( ob.Fun1() )
print ( ob.Fun2() )
```
