# SET OPERATION 

```python
S1 = {1,2,3,4}
S2 = {3,4,5,6}
S3 = {1,2}
```

```python
print (S1.union (S2))
print (S1 | S2)
```

```python
print (S1.intersection (S2))
print (S1 & S2)
```

```python
print (S1.difference (S2))
print (S1 - S2)
```

```python
print (S1.symmetric_difference(S2))
print (S1 ^ S2)
```

```python
print (S3.issubset )
```
