# List create in easy Way 

```python
L1 = []
for i in range (1,11) :
    L1.append (i * 2)
L1
```

```python
L2 = [ i*2 for i in range (1 , 11 )]
L2
```

# with if Statement 

```python
L1 = []
for i in range (1,11) :
    if i % 2 == 0 :
        L1.append (i)
L1 
```

```python
L3 = [x for x in range (11) if x % 2 == 0 ]
L3
```

# SET Comprehensive 

```python

S1 = {x for x in range (1,11)}
S1
```

```python
type (S1)
```

# Dict Comprehensive 

```python
D1 = {x : y for x in range (1,11) for y in range (10,21)}
D1
```

# Advanced IF Alse with List 

```python
L4 = [ "EVEN" if x%2 == 0 else "ODD" for x in range (11)]
L4
```

# Nested Loop

```python
L5 = [ (x,y) for x in range (3) for y in range (2) ]
L5
```

# List with Function 

```python
L4 = ['python' , 'java' , 'javascript' , 'PHP']
L5 = [word.upper() for word in L4]
L5
```

```python
L6 = [word for word in L4 if word.islower()]
L6
```

# Nested list with comprehensive

```python
L7 = [ [1,2] , [3,4] , [5,6] ]
L8 = [ x for i in L7 for x in i ]
L8
```

# List Comptrehensive with Set and Dictionary Comprehensions


```python
L9 = {x for x in [1,2,2,2,4,6,6]}
L9
```

```python
L10 = {x : x*x for x in range (1, 11)}
L10 
```
