# JSON = Javascribt object notation

```python
import json 
```

# dumps and loads

```python
DATA = {
"Name" : "Sushil",
"Rollno": 11 ,
"Age" : 20 ,
"cource" : "CE" 
}
```

```python
S = json.dumps (DATA )
S
```

```python
S1 = json.dumps (DATA ,indent= 4)
S1
```

```python
type (S) # JSON String
```

```python
Data = json.loads (S)
Data
```

# dump and load

```python
with open ("demo.json" , "w") as F : 
    json.dump (Data , F , indent = 10)
```

```python
with open ("demo.json" ,"r") as F :
    Data = json.load (F)
    
```
