# formating format ()

```python
name = "Sushil zangade"
age = 20 
print ("My Name = {} and My Age = {}".format (name , age))
```

```python
print ("My Name = {0} and My Age = {1}".format (name , age))
```

```python
print ("My Name = {name} and My Age = {age}".format (name = 'Mukesh', age= 50))
```

# Fstring [ Python 3.6 ]

```python
print (f"My Name = {name} and My Age = {age}" )
```

# Set Allignment to Text 

```python
print ( f"{'Hellow' : <10} " ) # Left
print ( f"{'Hellow' : >10} " ) # Right 
print ( f"{'Hellow' : ^10} " ) # Center
```
