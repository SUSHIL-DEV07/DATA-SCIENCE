# Operation in binary file

```python
import pickle

Student = {
    "name" : "Sushil",
    "Rollno" : 23 
}

F = open ("A.jpg" ,"wb") 
pickle.dump (Student , F) 
F.close ()
```

```python

F = open ("A.jpg" ,"rb")
pickle.load (F)
```
