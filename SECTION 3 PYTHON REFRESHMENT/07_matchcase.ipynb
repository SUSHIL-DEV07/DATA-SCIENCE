```python
code = 1000 
```

```python
match code : 
    case 404 : 
        print ("Page Not Found") 
    case 502 : 
        print ('Error 502')
    case 199 :
        print ('Error 199 ')
    case _ : 
        print ("Invalid code")
```

```python
POINT = (0 , 0)

match POINT: 
    case (0 , 0) :
        print ("Origin")
    case (0, y) :
        print (f"On Y axis : {y}") 
    case (x , 0 ):
        print (f'On X axis : {x}') 
    case (x , y) :
        print (f'ON X = {x} and Y = {y} axis ')
```
