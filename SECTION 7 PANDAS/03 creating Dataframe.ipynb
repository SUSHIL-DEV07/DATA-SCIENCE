# create Dataframe using list

```python
import pandas as pd 
```

```python
data = [["Sushil" , 91],
       ["Rohit" , 99] ,
       [ "Ganesh", 71],
       ["yash" , 55]] 
data
```

```python
df = pd.DataFrame (data , columns = ["Name" , "Marks"])
df
```

# Create Dataframe from dict and list 

```python
D2 = pd.DataFrame ({
    'name' : ['Sushil' , 'Alisa' , 'Mishra'] ,
    'marks' : [94,76,99] 
     })
D2 
```
