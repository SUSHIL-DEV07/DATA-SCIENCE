```python
D3 = df.loc [0:3 , ['Film','IMDB']] # Return Copy not view 
print ("Created DF = \n" , D3)
```

# At() to import single cell

```python
df.at [0 , 'Actor'] # It speed fast than loc 
```

```python
import time 
```

```python
start = time.time () 
df.loc [7 , "IMDB"] 
end = time.time () 

print ("loc time = " , end - start) 
```

```python
start = time.time () 
df.at [7 , "IMDB"] 
end = time.time () 

print ("at time = " , end - start) 
```
