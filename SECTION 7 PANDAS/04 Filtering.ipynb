```python
import pandas as pd
```

```python
Data = pd.read_csv ('Movie.csv') 
```

```python
print (Data.columns)
Data.columns = Data.columns.str.strip () 
print (Data.columns)

Data 
```

# Filter by condition

```python
Data [Data["IMDB"] >= 8]
```

```python
Data [Data['IMDB'] >= 7] ["Film"] # Only give name of Film 
```

```python
Data [ (Data['IMDB'] >= 8) & (Data["Year"] > 2017)  ] ["Actor"] # Single Column select 
```

```python
Data [ (Data['IMDB'] >= 8) & (Data["Year"] > 2017) ]  [ ["Actor","Film"] ] # Multiple column select 
```

# query () Function Easy filter 

```python
Data .query ("IMDB >= 7")
```

```python
Data.query ('IMDB >= 8 & Year >= 2017' )
```

```python
Data.query ("IMDB >= 8 and Year > 2017")[["Film" ,"IMDB" ,'Actor']]
```
