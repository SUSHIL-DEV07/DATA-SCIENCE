Question 8 (10 Marks)

Create:

arr = np.array([10,15,20,25,30,35,40])

Using np.where():

Replace every value greater than 25 with 999.

```python
# wrong code 
A = np.array ([10,15,20,25,30,35,40])
cond = A > 25
A [cond] = 999 
A

# right code 
B = np.array ([10,15,20,25,30,35,40])
B = np.where (cond , 999 , A) 
B
```
