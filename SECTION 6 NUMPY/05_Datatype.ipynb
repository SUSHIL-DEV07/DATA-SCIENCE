```python
import numpy as np 
D1 = np.array ([1,2,3]) 
D1.dtype 
```

```python
D2 = D1.astype (np.int32)
D2.dtype 
```

```python
print ("D1 Size = ", D1.nbytes)
print ("D2 Size = ", D2.nbytes)
```

# String 

```python
A1 = np.array(["Apple" ,"Ball", "Cat"] )
print ("A1 Datatype =  ", A1.dtype) 
```

# Complex 128

```python
A2 = np.array ([1+2j , 3+4j , 5+6j])
print (A2.dtype)
A2.nbytes
```

# object

```python
A3 = np.array ([{'a' : 1}, "String" , [1,2,4] ] , dtype = "object")
A3.dtype 
```

# BOOL

```python
A4 = np.array ([True,False,True,False ,True]) 
A4.dtype
```
