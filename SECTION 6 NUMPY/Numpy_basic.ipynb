# Numpy store element in contingous memory block 

```python
import sys 

size = 1000 

L1 = list(range (size))
N1 = np.array(L1) 

print ("List Size : ", sys.getsizeof(L1) * len (L1) ," Bytes" )
print ("np array size : " , N1.nbytes , " Bytes")
```

# Vectorization – No More Loops!

 NumPy avoids loops by applying operations to entire arrays at once using SIMD (Single Instruction, Multiple Data) and other low-level optimizations. SIMD is a CPU-level optimization provided by modern processors.

```python
import time 
```

```python
# each element add 5 

start = time.time () 
L2 = [x + 5 for x in L1]
end = time.time () 

print ("List time : " , end - start )
```

```python
start = time.time () 
L3 = N1 + 5  # It possible because it store same Datatype 
end = time.time () 

print ("np array time : " , end - start ) 
```
