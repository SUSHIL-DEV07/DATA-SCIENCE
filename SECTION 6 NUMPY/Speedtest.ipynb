```python
import numpy as np 
import time 

size = 1_000_000
L1 = list (range (size) )
L2 = list (range (size) )
start = time.time() 

add = [x + y for x, y in zip (L1,L2)]
end = time.time () 

print (add [:10]) 
print ("Time : ", end - start )
```

```python

L1 = np.array (L1)
L2 = np.array (L2)

S1 = time.time () 
L3 = L1 + L2
E1 = time.time () 

Time = E1 - S1


print (L3 [:10])
print ("Time : ",Time) 
```
