```python
import matplotlib.pyplot as plt 

plt.style.use ("petroff10") 

plt.show () 
```

```python
l = ["Sachin" , "sewag" ,'kohli' , "Youraj"] 
runs = [18000 , 8000 , 10000 , 12500]

plt.pie (runs , labels = l ) 
```

# Change color 

```python
c = ["Black","green","Blue","red"]
plt.pie (runs , labels = l , colors = c) 
```

# add edge and style 

```python
plt.pie (runs , colors = c , labels = l ,
        wedgeprops = {
    "edgecolor" : "black",
    "linewidth" : 2 ,
    "linestyle" : "-."  
        } ) 
```

# Highlight any area 

```python
e = [0.1 , 0 , 0 , 0 ]

plt.pie (runs , labels = l , colors = c , explode = e) 
```

# add shadow 

```python
plt.pie (runs , labels = l , colors = c , explode = e , shadow = True)  
```

# rotate piechart 

```python
plt.pie (runs , labels = l , colors = c , startangle = 90 )  
```

# add percentage 

```python
plt.pie (runs , labels = l , autopct = "%1.1f%%")  
```
