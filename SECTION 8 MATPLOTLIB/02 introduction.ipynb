```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot (years , runs )
```

```python

plt.xlabel ("Runs")
plt.ylabel ("Years")
plt.title ("Sachin Tedulkar Runs ")
plt.plot (years , runs )
plt.show () 
```

# Multiple lines in Single Graph

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot (years , kohli, label = "Virat Kholi" ) 
plt.plot (years , sehwag , label = "Viredra Sewag")

plt.xlabel ("Years")
plt.ylabel ("Runs")

plt.title("Runs Comparision") 
plt.legend () 
```

# Add marker and color Edit  (String Formating In Matplotlib)

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot (years , kohli,"r8:" , label = "Virat Kholi" ) 
plt.plot (years , sehwag ,"g_-.", label = "Viredra Sewag")

plt.xlabel ("Years")
plt.ylabel ("Runs")

plt.title("Runs Comparision") 
plt.legend () 
```

# without String Formating 

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot (years , runs , color = "red" , linestyle = "--" , label= "Sachin")
plt.plot (years , kohli , color = "blue" , linestyle = '-.' , label = "Kohli")
plt.plot (years , sehwag , color = "black" ,  label = "sehwag")


plt.legend () 
```

# linewidth , grid , tight_Layout 

```python
plt.tight_layout ()
plt.grid  = True
plt.plot (years , kohli , "r--" ,linewidth = 10   ) 
```
