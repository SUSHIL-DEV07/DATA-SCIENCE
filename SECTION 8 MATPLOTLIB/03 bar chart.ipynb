```python
import matplotlib.pyplot as plt 
```

```python

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar (years , runs ) 
plt.xlabel ("Years")
plt.ylabel ("Runs")
plt.title ("Sachin Run") 
```

# bar chart with miltuple values 

```python
import numpy as np 

 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange ( len (years)) 
w = 0.25
```

```python

plt.bar (x + w, sachin , width = w , label = "Sachin") 
plt.bar (x , sehwag , width = w , label = "sehwag") 
plt.bar (x - w, kohli , width = w , label = "kohli")

plt.legend () 

plt.xticks (x , years ) # show actual years 
```

# Text on Bar 

```python

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800] 

plt.bar (players , runs )

for i in range (len (players) ):
    plt.text (i , runs [i] + 20 , str (runs [i]) , ha = "center") 
```
