```python
import json 

def load_data (filename) :
    with open (filename , "r") as F:
        return json.load (F)
        
```

```python
def pages_you_might_like ( id , data ) :
    
    all = {}

    for user in data ['users'] : 
        all [user ["id"]] = set (user ['liked_pages'])
        #{1: {101}, 2: {102}, 3: {101, 103}, 4: {104}}    

    if id not in list(all.keys()) :
        return 

    self = all[id] 
    # {101}
    sugg = {}

    for userid , pages in all.items() : 
        
        if id != userid : 
            shared_page = self.intersection(pages)
    
            for page in pages : 
                if page not in self :
                    sugg[page] = sugg.get(page,0) + len(shared_page)

    sorted_sugg = sorted (sugg.items (),key = lambda x : x[1] , reverse = True)
    return ([id for id ,_ in sorted_sugg])
            #     pass
        
```

```python
data = load_data ("messive_data.json")
userid = 1 
ref = pages_you_might_like (userid , data)
print (f"ID : {userid} page suggession : {ref}")
```
