```python
import matplotlib.pyplot as plt 
```

```python
plt.plot ([1,2,3] , [4,5,6] )
plt.show () 
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot (years , runs ) 
plt.show () 
```

```python
plt.plot (years,runs)
plt.xlabel ("Years") 
plt.ylabel ("runs") 
plt.title ("Sachin Runs") 
plt.show ()
```

# Multiple plot 

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot (years, runs , label = "Sachin")
plt.plot (years , kohli , label = 'Kohli')
plt.plot (years , sehwag, label = 'Sehwag')

plt.legend ()
plt.show () 
```

# add Legend 

```python
plt.plot (years , runs)
plt.plot (years , kohli)

plt.legend (["Sachin" , "Kohli"])
plt.show () 
```

# format String 

```python
plt.plot (years , runs ,"r.--")
plt.plot (years , kohli , "b_--")

plt.legend (["Sachin" , "kohli"])
plt.show ()
```

# line change 

| Character | Description |
|-----------|-------------|
| `'-'` | Solid line style |
| `'--'` | Dashed line style |
| `'-.'` | Dash-dot line style |
| `':'` | Dotted line style |

```python
plt.plot (years , runs , "o:")
plt.plot () 
```

# Color

| Character | Color |
|-----------|-------|
| `'b'` | Blue |
| `'g'` | Green |
| `'r'` | Red |
| `'c'` | Cyan |
| `'m'` | Magenta |
| `'y'` | Yellow |
| `'k'` | Black |
| `'w'` | White |

```python
plt.plot (years , runs ,"k") 
plt.plot ()
```

# change color , line style , tight_layout , linewidth , add grid 

```python
plt.plot (years , runs , color = "black" , linestyle = "--" , label = "Sachin" ) 
plt.plot (years , kohli , color = "RED" , linestyle = ":" , label = "kohli" , linewidth = 3)

plt.tight_layout ()
plt.grid () 
plt.plot ()
```
