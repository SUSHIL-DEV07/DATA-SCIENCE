```python
import numpy as np
```

```python
A = np.array ([[1,2,3] ,
              [4,5,6]])
A.dtype 
A.astype ("str").dtype
```

```python
B = A.copy ()
B = B.astype ("int32") 
print ("A = ", A.nbytes ) 
print ("B = ", B.nbytes ) 
```

```python
C = np.array (["S1" , "S2" ,"S3"] , dtype = "U10")
print (C)
print (C.dtype) 
```

```python
D = np.array ([1+1j , 2+2j ,3+3j] )
D.dtype
D.nbytes
```

```python
E = np.array ([ 1,2 , {1,2} , {1,2} , {"1" :10 }])
E.dtype
E[2]
```
