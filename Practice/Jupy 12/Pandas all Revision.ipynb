# Lecture 7 melt and pivot 

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

Data = pd.DataFrame(data)

# Display the DataFrame
print(Data)
Data.columns = Data.columns.str.strip()
```

```python
Data = Data.melt (id_vars = "Name" , value_vars = ["Math","Science","English"] , var_name = "Subject" , value_name = "Score") 
```

```python
Data.pivot (index = "Name" , columns = "Subject" , values = "Score" ) 
```

```python
Data 
```

```python
Data.loc [9] = ["Alice" , "MAth" , 99]
```

```python
Data
```

```python
Data.pivot_table (index = "Name" , columns = "Subject" , values = "Score" , aggfunc = "mean") 
```

# Lecture 8 groupby with aggfunc

```python
Data = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
```

```python
Data
```

```python
Data.groupby (["Department" , "Gender"])["Salary"].mean() 
```

```python
Data.groupby ("Team")["Salary"].agg (['mean' , 'max']) 
```

```python
Data.groupby ("Team")["Salary"].agg (avg_Sal = "mean" ,
                                    max_sal = "max")
```

```python
Data.groupby ("Team").agg ({
    "Salary" : "mean",
    "Age" : "max"
}) 
```

```python
Data ["Team Salary"] =  Data .groupby('Team') ["Salary"] .transform ("mean") 
```

```python
Data
```

```python
Data.groupby ("Department") .filter (lambda x : x["Salary"].mean () != "M") 
```

# Lecture 9 merge and concat

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees 
```

```python
departments
```

```python
pd.merge (employees , departments , on = "DeptID") 
```

```python
pd.merge (employees , departments , on = "DeptID",  how= "outer")
```

```python
df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"]})
 
```

```python
pd.concat ([df1,df2])
```

```python
pd.concat ([df1 , df2 ],axis = 1) 
```
