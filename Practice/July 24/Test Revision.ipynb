# np.where (condition ,x, y , Array)

```python
A = np.array ([10,20,30,40,50,60])
np.where (A>30 , "Yes" , "No")
```

```python
A = np.array ([35 ,25,100,99])
Result = np.where (A >= 35 , "PASS" ,"Fail") 
Result
```

```python
A = np.array ([-1,10,15,-4])
A = np.where (A < 0 , 0 , A) 
A
```

# Syntax 2 np.where (condition)

```python
A = np.array ( [10,20,30,40,50] )
np.where (A > 30 )
```

```python
A = np.array ([[10,20,30] ,
             [40,50 ,20]])
B , C = np.where (A < 30 )
print ("B = " , B)
print ("C = " , C)
```
